```python
import csv
csv_columns=['name','age','address']
dict=[
    {'name':'mohini','age':23,'address':'kathmandu'},
    {'name':'moon','age':22,'address':'kathmandu'}
]
csvfile=open("data.csv",'w')
writer=csv.DictWriter(csvfile,fieldnames=csv_columns)
writer.writeheader()
for data in dict:
    writer.writerow(data)
csvfile.close()
```
