```python
# return type funtion
def hello(x):
    return x
y = hello("hello world")
print(y)
```

```python
def area(x,y):
    a=x*y
    return a
ar=area(10,2)
print("the area is =",ar)
```

```python
def cal(x,y,z):
    a=x*y
    v=a*z
    return a,v
ar=cal(10,2,2)
area,volume=ar
print("the area is =",area)
print("the volume is=",volume)
```

```python
# WAP to make a calculator by using return type function
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b
a=int(input("enter a="))
b=int(input("enter b="))
o=input("enter + - * /")
if o == '+':
    print(add(a,b))
elif o == '-':
    print(sub(a,b))
elif o == '*':
    print(mul(a,b))
elif o == '/':
    print(div(a,b))
else:
    print("invalid operator")


```
